```python
import pandas as pd
import re
import os
import liwc
import jieba
import pickle
from ltp import LTP
from ltp import StnSplit
```

```python
ltp_model_path = 'ltp_base'
# 初始化LTP模型
ltp = LTP(ltp_model_path)
```

```python
def modifier_ratio(words):
    modifier_count = 0
    result = ltp.pipeline(words, tasks=['cws',"pos"])
    pos_tags = result.pos[1]
    for pos in pos_tags:
        if pos=="a":  # 形容词标签以"a"开头
            modifier_count += 1
        elif pos=="d":  # 副词标签以"d"开头
            modifier_count += 1
    return modifier_count / len(pos_tags[0])
```

```python
ratio = modifier_ratio('咖喱很好吃的,尤其是温泉蛋咖喱,就是麻婆茄子的量太少,不管是多大份量都是一样的')
```

```python
dict = pd.read_csv("LIWC2015 Dictionary - Chinese (Simplified).dicx")

def count_liwc(words,categories):
    counts = {cat: 0 for cat in categories}
    word_aim = []
    for word in words:
        if word in dict.index:
            for cat in categories:
                if not pd.isna(dict.loc[word, cat]):
                    counts[cat] += 1
                    word_aim.append(word)
    total = 0
    for value in counts.values():
        total += value
    return total, counts, set(word_aim)
```
